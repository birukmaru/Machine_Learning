```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('sat_gpa_data.csv')
```

```python
X = df['SAT'].values.reshape(-1, 1) 
y = df['GPA'].values
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
predicted_gpa = model.predict(X)

m = model.coef_[0]
c = model.intercept_

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")
```

```python
mse = mean_squared_error(y, predicted_gpa)
r2 = r2_score(y, predicted_gpa)

print(f"Mean Squared Error: {mse}")
print(f"R-squared (R²): {r2}")
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, predicted_gpa, color='red', label='Regression Line')
plt.xlabel("SAT Score")
plt.ylabel("GPA")
plt.title("Linear Regression: SAT vs GPA (Using Scikit-learn)")
plt.legend()
plt.show()
```
