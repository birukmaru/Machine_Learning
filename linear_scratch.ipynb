```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sat_gpa_data.csv')
```

```python
X = df['SAT'].values.reshape(-1, 1)
Y = df['GPA'].values
```

```python
mean_X = np.mean(X)
mean_y = np.mean(Y)
n = len(X)
```

```python
numerator = np.sum((X - mean_X) * (Y - mean_y))
denominator = np.sum((X - mean_X) ** 2)
```

```python
m = numerator / denominator

c = mean_y - m * mean_X

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")
```
